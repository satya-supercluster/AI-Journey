### USED WHEN WE HAVE DIFFERENT SHAPES

```python
import numpy as np

a = np.array([1,2,3])
b = np.array([4])   # Broadcasting means [4,4,4]

print(np.add(a,b))
```

#### Compatibility -> Same or one of them 1

```python
a = np.array([1,2,3])   # [1,2,3] -> [[1,2,3]] -> [[1,2,3],[1,2,3]]
# a->(3)->(1,3)->(2,3)
b = np.array([[4],[5]]) # [[4],[5]] -> [[4,4,4],[5,5,5]]
# b->(2,1)->(2,3)

# a->(1,3)
# b->(2,1)
# Compatible

# a->(1,3,5,6,1,1,9)
# b->(2,1,5,1,5,7,9)
# Compatible

# a->(1,3,5,6,1,1,9)
# b->(2,1,5,3,5,7,9)
# Not Compatible

print(np.add(a,b))
```

```python
a = np.random.random((1,3,5,6,1,1,9))
b = np.random.random((2,1,5,1,5,7,9))

print((np.add(a,b)).shape)
print(np.add(a,b))
```
